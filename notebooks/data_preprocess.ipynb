Classification task:
1. Melanoma
2. Melanocytic nevus
3. Basal cell carcinoma
4. Actinic keratosis
5. Benign keratosis (solar lentigo / seborrheic keratosis / lichen planus-like keratosis)
6. Dermatofibroma
7. Vascular lesion
8. Squamous cell carcinoma
9. None of the others

```python
cat_names=['melanoma','nevus','basal cell carcinoma','actinic keratosis','benign keratosis','dermatofibroma',
          'vascular lesion','squamous cell carcinoma','others']
cat_names_include = ['melanoma','melanoma metastasis','nevus','basal cell carcinoma','actinic keratosis','solar lentigo',
                     'seborrheic keratosis','lichenoid keratosis','dermatofibroma',
          'vascular lesion','squamous cell carcinoma']
df['melanoma'] = df['diagnosis'].apply(lambda x: 1 if (x == 'melanoma') or (x == 'melanoma metastasis') else 0)
df['nevus'] = df['diagnosis'].apply(lambda x: 1 if (x == 'nevus') else 0)
df['basal cell carcinoma'] = df['diagnosis'].apply(lambda x: 1 if (x == 'basal cell carcinoma') else 0)
df['actinic keratosis'] = df['diagnosis'].apply(lambda x: 1 if (x == 'actinic keratosis') else 0)
df['benign keratosis'] = df['diagnosis'].apply(lambda x: 1 if (x == 'solar lentigo') or (x =='seborrheic keratosis') or (x == 'lichenoid keratosis') else 0)
df['dermatofibroma'] = df['diagnosis'].apply(lambda x: 1 if (x == 'dermatofibroma') else 0)
df['vascular lesion'] = df['diagnosis'].apply(lambda x: 1 if (x == 'vascular lesion') else 0)
df['squamous cell carcinoma'] = df['diagnosis'].apply(lambda x: 1 if (x == 'squamous cell carcinoma') else 0)
df['others'] = df['diagnosis'].apply(lambda x: 1 if (x not in cat_names_include) else 0)
```

```python
df = df[['isic_id','age_approx','benign_malignant','sex','patient_id']+cat_names]
```

```python
df['disease'] = df[cat_names[0]]
for i in range(len(cat_names)):
    df.loc[df[cat_names[i]] == 1, 'disease'] = cat_names[i]

# df.loc[df['disease'].isna(), 'disease'] = 'others'
```

## dataset spliting
requirement: P(D_k|A) is the same in training/validation/testing set 

```python
import random
random.seed(2023)
train_perc,valid_perc,test_perc = 0.6,0.2,0.2
```

```python
train_set_ids=[]
valid_set_ids=[]
test_set_ids=[]
```

```python
for each_dis in cat_names:
    male_list = df.loc[((df['disease'] == each_dis) & (df['sex'] == 'male'))]['isic_id'].to_list()
    print('Group{{male,{}}}:{}'.format(each_dis,len(male_list)))
    female_list = df.loc[((df['disease'] == each_dis) & (df['sex'] == 'female'))]['isic_id'].to_list()
    print('Group{{female,{}}}:{}'.format(each_dis,len(female_list)))
    
    train_num_male = int(len(male_list)*train_perc)
    valid_num_male = int(len(male_list)*valid_perc)
    test_num_male = len(male_list) - train_num_male - valid_num_male
    
    train_num_female = int(len(female_list)*train_perc)
    valid_num_female = int(len(female_list)*valid_perc)
    test_num_female = len(female_list) - train_num_female - valid_num_female
    
    
    train_set_tmp = random.sample(male_list,k=train_num_male)
    valid_set_tmp = random.sample(list(set(male_list) - set(train_set_tmp)),k=valid_num_male)
    test_set_tmp = list(set(male_list) - set(train_set_tmp)-set(valid_set_tmp))
    
    train_set_ids.extend(train_set_tmp)
    valid_set_ids.extend(valid_set_tmp)
    test_set_ids.extend(test_set_tmp)
    
    # female
    train_set_tmp = random.sample(female_list,k=train_num_female)
    valid_set_tmp = random.sample(list(set(female_list) - set(train_set_tmp)),k=valid_num_female)
    test_set_tmp = list(set(female_list) - set(train_set_tmp)-set(valid_set_tmp))
    
    train_set_ids.extend(train_set_tmp)
    valid_set_ids.extend(valid_set_tmp)
    test_set_ids.extend(test_set_tmp)
```

```python
df['split'] = df['isic_id'].apply(lambda x: 'train' if x in train_set_ids else ('valid' if x in valid_set_ids else 'test'))
```
